```python
import pandas as pd
from newspaper import Article
import nltk

articles = pd.read_csv('newsArticlesWithLabels.tsv', sep='\t')

democrat_opinion = articles.loc[:, "democrat.vote"]

republican_opinion = articles.loc[:, "republican.vote"]

article_url = articles.loc[:, "url"]

democrat_num = []
republican_num = []
def numberSetter(dataframe, array):
    for i in dataframe:
        if(i == "Positive"):
            array.append(1)
        
        if(i == "SomewhatPositive"):
            array.append(0.5)
        
        if(i == "Negative"):
            array.append(1)
        
        if(i == "SomewhatNegative"):
            array.append(-0.5)
        
        if(i == "Neutral"):
            array.append(-1)
    return array
        

    
democrat_num = numberSetter(democrat_opinion, democrat_num)
republican_num = numberSetter(republican_opinion, republican_num)

#adding article urls to designated files
# d = open("Democrat.txt", "a")
# r = open("Republican.txt", "a")
# for i in range(articles.shape[0]):
#     if(democrat_num[i]-republican_num[i]>0):
#         d.write(article_url[i] + "\n")
        
#     else:
#         r.write(article_url[i] + "\n")
        
# d.close()
# r.close()
# len(democrat_num)
# len(republican_num)
# r = open("republican.txt", "a")
# r.write("\n")
# r.close()

# r = open("democrat.txt", "a")
# r.write("\n")
# r.close()



```
